```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r"E:\FRT\chromedriver.exe")
```

```python
url = "https://www.monsterindia.com/srp/results?query=software%20developer&searchId=7d0a3472-a4ae-4f22-a03e-fba23d3de99a"
```

```python
driver.get(url)
```

```python
job_titles=[]
company_names=[]
```

```python
import time
```

```python
for i in range(0,10):
    job_tags=driver.find_elements_by_xpath("//h3[@class='medium']")
    for i in job_tags:
        job_titles.append(i.text)
        
    companies_tags=driver.find_elements_by_xpath("//span[@class='company-name']")
    for j in companies_tags:
        company_names.append(j.text)
    driver.find_element_by_xpath("//button[@class='btn-next-prev btn-next']").click()
    time.sleep(3)
```

```python
job_title=job_titles[0:450]
job_titles
```
