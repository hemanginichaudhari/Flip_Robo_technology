```python
page=requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
odi_teams=[]
odi_matches=[]
odi_points=[]
odi_ratings=[]
```

```python
teams=soup.find_all('span',class_="u-hide-phablet")
teams
```

```python
for i in teams:
    odi_teams.append(i.get_text().replace("\n",""))
    
odi_teams[0:10]
```

```python
matches=soup.find_all('td',class_="table-body__cell u-center-text")
matches
```

```python
for i in matches:
    odi_matches.append(i.get_text().replace("\n",""))
    
odi_matches[0:10]
```

```python
points=soup.find_all('td',class_="table-body__cell u-center-text")
points
```

```python
for i in points:
    odi_points.append(i.get_text().replace("\n",""))
    
odi_points
```

As we see above matches and points having same scraping tag so both data is scraped in alternate position. so, separating that by slicing.

```python
odi_match = odi_points[1::2] + odi_points[::2] 
odi_match[0:9]
```

```python
odi_pts = odi_points[::2] + odi_points[1::2] 
odi_pts[0:9]
```

```python
ratings=soup.find_all('td',class_="table-body__cell u-text-right rating")
ratings
```

```python
for i in ratings:
    odi_ratings.append(i.get_text().replace("\n",""))
    
odi_ratings
```

```python
Odi_teams=odi_teams[0:9]
Odi_matchesh=odi_match[0:9]
Odi_points=odi_pts[0:9]
Odi_ratings=odi_ratings[0:10]
```

```python
import pandas as pd
ODI_Womens_data=pd.DataFrame({})
ODI_Womens_data['Teams']=Odi_teams
ODI_Womens_data['Matches']=Odi_matchesh
ODI_Womens_data['Points']=Odi_points
ODI_Womens_data['Ratings']=Odi_ratings
```

```python
ODI_Womens_data
```

```python
ODI_Womens_data.to_csv()
```
