```python
# importing the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
header_tags=[]
```

```python
header=soup.find_all('h2',class_='mp-h2')
header
```

```python
for i in header:
    header_tags.append(i.get_text().replace("\n",""))

header_tags
    
```

```python
import pandas as pd
header_data=pd.DataFrame({})
header_data['header']=header_tags
```

```python
header_data
```

```python
header_data.to_csv()
```
