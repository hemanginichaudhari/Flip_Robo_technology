```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r"E:\FRT\chromedriver.exe")
```

```python
url = "https://www.monsterindia.com/srp/results?query=Data%20Scientist&locations=new%20delhi&searchId=21842756-c51a-4cd4-adaa-e2907572ba50"
```

```python
driver.get(url)
```

```python
job_titles=[]
company_names=[]
```

```python
import time
```

```python
for i in range(0,10):
    titles_tags=driver.find_elements_by_xpath("//h3[@class='medium']")
    for i in titles_tags:
        job_titles.append(i.text)
        
    companies_tags=driver.find_elements_by_xpath("//a[@class='under-link']")
    for j in companies_tags:
        company_names.append(j.text)
    driver.find_element_by_xpath("//button[@class='btn-next-prev btn-next']").click()
    time.sleep(3)
```

```python
job_titles
```

```python
company_names
```

```python
Job_titles=job_titles[0:185]
```

```python
print(len(Job_titles),len(company_names))
```

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['title']=Job_titles
jobs['company']=company_names
```
